```python
#import image via cv2

import cv2
import os

def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images

data = load_images_from_folder('/Users/davidng/Documents/David/images')
```

```python
import json
with open('/users/jay/desktop/metadata.json') as f:
    metadata_dict = json.load(f)
```

```python
image_name=list(metadata_dict)
```

```python
#transform the points into bounding box

n_list=[]
bound_list=[]
new_list=[]
for i in image_name:
    num_points=len(metadata_dict[i]['bounds_x_y'])
    x=[]
    y=[]
    for j in range(num_points):
        x.append(metadata_dict[i]['bounds_x_y'][j]['x'])
        y.append(metadata_dict[i]['bounds_x_y'][j]['y'])
    x_max=max(x)
    x_min=min(x)
    y_max=max(y)
    y_min=min(y)
    bounds=(x_min,y_min,x_max,y_max)
    bound_list.append(bounds)
    new=(i,bounds)
    new_list.append(new)
print(new_list)
```
